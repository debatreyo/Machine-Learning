## `Gr Liv Area` Feature

>**`Gr Liv Area`**: Above grade (ground) living area square feet

```python
# visualize correlation between Ground living area of house and sale price
plt.figure(figsize=(6,4), dpi=220)

sns.scatterplot(x="Gr Liv Area", y="SalePrice", data=df,
           color="#085786", alpha=0.7)

plt.grid(linestyle=":", alpha=0.75)
plt.title("Ground Living Area v/s Sale Price")

plt.tight_layout();
```

### Observations
Visualization again reveals some **outlier houses** with relatively **lower `Sale Price`** despite higher **`Gr Liv Area`**.

### Outlier Houses

```python
# DF capturing outliers visualized above
mask = (df["SalePrice"] < 200000) & (df["Overall Qual"] > 8)
outlier_houses = df[mask]

outlier_houses
```

For **training a ML model** such **outliers** can produce **incorrect results**.
>_Model may not generalize well if trained on dataset with many such outliers._

### Conclusion
Need to **remove** such outlier examples before feeding data to ML model for training.

## Remove outliers

```python
# drop outliers

# filter outliers
mask = (df["Gr Liv Area"] > 4000) & (df["SalePrice"] < 300000)
# retrieve index values of outliers
index_to_drop = df[mask].index
```

```python
df_NoOutlier = df.drop(labels=index_to_drop, axis=0)
```

```python
# new df with outlier houses removed
df_NoOutlier.head()
```

### Visualizing data without the outliers

```python
# visualize correlation between Ground living area of house and sale price
plt.figure(figsize=(6,4), dpi=220)

sns.scatterplot(x="Gr Liv Area", y="SalePrice", data=df_NoOutlier,
           color="#169480", alpha=0.7)

plt.grid(linestyle=":", alpha=0.75)
plt.title("Ground Living Area v/s Sale Price")

plt.tight_layout();
```

```python
# visualize correlation between overall quality of house and sale price
plt.figure(figsize=(6,4), dpi=220)

sns.scatterplot(x="Overall Qual", y="SalePrice", data=df_NoOutlier,
           color="#169480", alpha=0.7)

plt.grid(linestyle=":", alpha=0.75)
plt.title("Overall Quality v/s Sale Price")

plt.tight_layout();
```
