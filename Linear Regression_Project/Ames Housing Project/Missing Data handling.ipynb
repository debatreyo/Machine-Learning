```python
def missing_percent(df):
    """
    computes % of missing values per feature
    in datafram in descending order.
    
    returns a Pandas Series
    
    NOTE: includes only those features
    where >0.00% values are missing
    
    Assumes:
    df is Pandas DataFrame
    """
    # no. of datapoints per feature
    # alternative = len(df)
    size = df.index.value_counts().sum()
    
    # total NaN values per feature (col)
    null = df.isnull().sum(axis=0)
    
    # percentage
    percentage = np.round((null / size)*100, decimals=2)
    
    # return only those features where >0% values missing
    return percentage[percentage > 0.00].sort_values(ascending=False)
```

```python
# percentage of missing data per feature
nan_percent = missing_percent(df)
nan_percent.index
```

```python
# visualizing number missing values per feature
fig = plt.figure(figsize=(9, 5), dpi=220)

sns.barplot(x=nan_percent.index, y=nan_percent.values,
           palette="tab20")
plt.xticks(rotation="vertical")

plt.grid(linestyle=":", linewidth=0.5, alpha=0.7)
plt.axhline(y=10, linestyle="--", color="#120063",
            label="Missing 10% Data", alpha=0.8)
plt.axhline(y=75, linestyle="--", color="#7E2A10",
            label="Missing 75% Data", alpha=0.8)
plt.axhline(y=1, linestyle=":", color="#000000",
            label="Missing 1% Data", alpha=0.8)

plt.ylabel("Percentage of Missing values")
plt.xlabel("\nFeatures")
plt.legend(loc="upper right")

plt.title("Missing Data per Feature")
plt.tight_layout();
```

### Dealing with Features related to Basement



```python
# numeric columns -> will replace NaN with 0 (i.e does not have basement)

basement_numeric_cols = [
    "BsmtFin SF 1", "BsmtFin SF 2",
    "Bsmt Unf SF", "Total Bsmt SF",
    "Bsmt Half Bath", "Bsmt Full Bath"
    
]

# categorical columns -> will replace NaN with "None" (i.e. does not have basement)

basement_categoric_cols = [
    "Bsmt Qual", "Bsmt Cond",
    "Bsmt Exposure", "BsmtFin Type 1",
    "BsmtFin Type 2"
]
```

```python
# replace basement categorical features NaN values

df[basement_categoric_cols] = df[basement_categoric_cols].fillna(value="None")
```
