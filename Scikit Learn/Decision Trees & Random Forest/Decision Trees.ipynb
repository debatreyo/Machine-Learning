````python
# helper function

def missing_percent(df):
    """
    computes % of missing values per feature
    in datafram in descending order.
    
    returns a Pandas Series
    
    NOTE: includes only those features
    where >0.00% values are missing
    
    Assumes:
    df is Pandas DataFrame
    """
    # no. of datapoints per feature
    # alternative = len(df)
    size = df.index.value_counts().sum()
    
    # total NaN values per feature (col)
    null = df.isnull().sum(axis=0)
    
    # percentage
    percentage = np.round((null / size)*100, decimals=2)
    
    # return only those features where >0% values missing
    return percentage[percentage > 0.00].sort_values(ascending=False)
````

````python
nan_percent = missing_percent(df)

nan_percent
````

# Setting up the data

````python
# input features
X = df.drop(labels="species", axis=1)
# target variable
y = df["species"]
````

````python
# one-hot encoding categorical input features

# drop_first=True -> for avoiding unnecessarily increasing no. of features
# Ex. if category has 2 possible values A & B -> Encoding A as 0/1
# atumoatically captures information on B
X = pd.get_dummies(data=X, drop_first=True)
````

### NOTE
(_Unlike other algorithms_) For **Decision Tree** since **only one feature is being looked at at any instance (`node`)** for splitting (_deciding_),
**feature scaling is NOT required**.

## Creating Training set and Test set

````python
# for splitting dataset

from sklearn.model_selection import train_test_split
````

````python
# training set = 70%; test set = 30%

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=0)
````

## Building a Decision Tree Classifier Model

````python
# import model

from sklearn.tree import DecisionTreeClassifier
````

````python
# create model with default hyperparameters
# will decide optimal hyperparameters later by experimentation

model = DecisionTreeClassifier()
````

### Training on the Training set

````python
# fit training data to model
model.fit(X=X_train, y=y_train)
````

### Predict on Test set

````python
# predict species
test_predictions = model.predict(X=X_test)
````

````python
print(f"First 5 Species predictions on Test set:\n\n{test_predictions[0:5]}")
print(f"First 5 Species (Actual values) on Test set:\n\n{y_test.head(5)}")
````

### Evaluating the Trained Decision Tree Classifier model

````python
# metrics for evaluation

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
````

````python
# evaluation report

report = classification_report(y_true=y_test, y_pred=test_predictions)

print(report)
````

````python
cm = confusion_matrix(y_true=y_test, y_pred=test_predictions,
                     labels=model.classes_)
````

````python
# visualize confusion matrix

cm_display = ConfusionMatrixDisplay(confusion_matrix=cm,
                      display_labels=model.classes_)
````

````python
fig, ax = plt.subplots(figsize=(3,3), dpi=220)
cm_display.plot(ax=ax, im_kw={"cmap":"Blues"})

fig.suptitle("Confusion Matrix")
````

### Helper Function to Generate Model's report & Tree diagram

Will be helpful when **tuning model's hyperparameters** and **comparing performance**.

````python
def report_model(model, toPrint=False):
    """
    Does the following:
    
    1. Finds predictions by a Trained Decision Tree Classifier
    2. Prints a Classification Report for that model's predictions
    3. if toPrint=True, Plots the diagram of entire Decision Tree in action
    
    Assumes:
    model -> Trained Decision Tree Classifier model
    """
    # predict on test set
    predictions = model.predict(X_test)
    # generate classification report for the predictions
    report = classification_report(y_true=y_test,
                                                  y_pred=predictions)
    
    # classification report
    print(f"Classification Report:\n\n{report}\n\n")
    
    if toPrint:
        # create tree diagram
        plt.figure(figsize=(12, 8), dpi=220)

        plot_tree(decision_tree=model, feature_names=list(X.columns),
             filled=True)

        plt.title(f"Decision Tree Classifier in Action")
        plt.tight_layout();
````

````python
# test helper function

report_model(model, toPrint=True)
````

# Hyperparameter Tuning

## Maximum Tree depth

````python
# set max tree depth
model_shallower = DecisionTreeClassifier(max_depth=2)
````

````python
# fit data to model
model_shallower.fit(X=X_train, y=y_train)
````

````python
# generate model's report

report_model(model_shallower, toPrint=True)
````

### Observation
1. Decision Tree is now **shorter & simpler** with **few features considered** for splitting examples (_less no. of levels in tree_)
2. **Performance** has **deteriorated** compared to when tree depth was not constrained
    - (_unconstrained_) `f-1 scores = 0.98, 0.94, 1.00`
    - (_constrained_) `f-1 scores = 0.96, 0.85, 0.98`

## Maximum Leaf nodes

````python
# contrain maximum leaf nodes
model_maxLeaf = DecisionTreeClassifier(max_leaf_nodes=3)
````

````python
# fit model to training set
model_maxLeaf.fit(X=X_train, y=y_train)
````

````python
# generate model's report

report_model(model_maxLeaf, toPrint=True)
````

### Observation
1. Decision Tree is now **shorter & simpler** with **few features considered** for splitting examples (_less no. of nodes_)
2. **Performance** has **deteriorated** compared to when tree depth was not constrained
    - (_unconstrained_) `f-1 scores = 0.98, 0.94, 1.00`
    - (_constrained_) `f-1 scores = 0.96, 0.85, 0.98`
3. `gini coefficient` is **NOT zero** for any node. (_NO pure node_)

## Criterion for deciding split (_Default: Gini coefficient_)

````python
# create model with Information gain as criteria
model_entropy = DecisionTreeClassifier(criterion="entropy")
````

````python
# fit model to training set
model_entropy.fit(X=X_train, y=y_train)
````

````python
# generate model's report

report_model(model_entropy, toPrint=True)
````

### Observation
1. Model **performs well** since it was **NOT constrained** in any way.

````python
# confusion matrix

# predictions
predict_entropy = model_entropy.predict(X=X_test)

cm_entropy = confusion_matrix(y_true=y_test, y_pred=predict_entropy,
                     labels=model_entropy.classes_)

# visualize confusion matrix

cm_display = ConfusionMatrixDisplay(confusion_matrix=cm_entropy,
                      display_labels=model_entropy.classes_)
fig, ax = plt.subplots(figsize=(3,3), dpi=220)
cm_display.plot(ax=ax, im_kw={"cmap":"Blues"})

fig.suptitle("Confusion Matrix")
````
