# Dataset
Experiment (5000 participants) conducted to study **effects of age & physical health on hearling loss** (ability to hear **high pitched tones**).

### Data (Experiment results)
**Age** of participant.\
Whether they passed an **audio test**.

### Input Features
>**`age`**- Age of participant (years)\
**`physical_score`**- Test score

### Target Label
> **`test_result`** $\rightarrow$
>> **`0 = Failed audio test`** \
>> **`1 = Passed audio test`**

### Data Source
[Kaggle Open-source Datasets](https://www.kaggle.com/datasets/prasenjitsharma/hearing-test-classification-problem)

```python
# for data manipulation & visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Data

```python
# filepath
filepath = r"D:\ML\Machine Learning_Practical\Logistic Regression_Project\Classifying Patients and Hearing ability\data\hearing_test.csv"

# dataframe
df = pd.read_csv(filepath)
```

# Exploratory Data Analysis- Understanding the dataset

```python
# dataset
df.head()
```

```python
# basic statistics

df.describe().transpose()
```

```python
# check if dataset is skewed

fig, ax = plt.subplots(figsize=(10,6), dpi=220)

sns.countplot(x="test_result", data=df, ax=ax,
             palette={0:"#AF2821", 1:"#519F1B"},
             label=["Failed test", "Passed test"])

ax.legend(labels=["Failed test", "Passed test"], loc=(1.02, 0))

ax.set_xticklabels(["Failed test", "Passed test"])

ax.grid(linestyle=":", linewidth=0.85, alpha=0.65)

ax.set_ylabel("No. of examples")
fig.suptitle("Checking Dataset for Skewness")
fig.tight_layout();
```

### Observation
1. Dataset is **not too skewed**
    - No. of examples with `0` labels and `1` labels are comparable

```python
# check distribution of data by each feature

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,6), dpi=220)
ax1, ax2 = ax
sns.boxplot(x="test_result", y="age", data=df, ax=ax1,
            palette={0:"#AF2821", 1:"#519F1B"})

sns.boxplot(x="test_result", y="physical_score", data=df, ax=ax2,
            palette={0:"#AF2821", 1:"#519F1B"})



ax1.set_xticklabels(["Failed test", "Passed test"])
ax1.grid(linestyle=":", linewidth=0.85, alpha=0.8)

ax2.set_xticklabels(["Failed test", "Passed test"])
ax2.grid(linestyle=":", linewidth=0.85, alpha=0.3)

fig.suptitle("Distribution of Test result\n(by Age & Physical score of participants)")
fig.tight_layout();
```

### Observation
1. **Median age** of participants who **did not pass audio test** are around **15 years older** than those who did pass
---
2. **Median physical condition (`physical_score`)** of participants who **did not pass audio test** are around **13 points lower** than those who did pass

>_Both observations align with common expectations around age and disability of human beings_.
>>_People who are **younger** and in **good physical condition** are much **more likely to pass the audio test** than the rest_.

#### Age & Physical score- Relationship

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=220)

sns.scatterplot(x="age", y="physical_score", data=df, ax=ax,
               hue="test_result", alpha=0.5,
               palette={0:"#AF2821", 1:"#519F1B"})

ax.grid(linestyle=":", linewidth=0.85, alpha=0.65)

fig.suptitle("Age (years) v/s Physical condition\n(0: Failed audio test, 1: Passed audio test)")
fig.tight_layout();
```

### Observation
1. Scatterplot clearly reveals the expected **trend** that people who are **younger** usually have **better physical condition** as indicated by **higher `physical_score`** parameter.

>_Observation matches the conclusions from previous observations_.

#### Combining above analysis into single visualization

```python
plt.figure(dpi=250)

sns.pairplot(data=df, hue="test_result",
            palette={0:"#AF2821", 1:"#519F1B"},
            height=5, plot_kws={'alpha': 0.6})

plt.legend(title='Test Results',
           labels=['0: Failed', '1: Passed'], loc=(1.05, 0.4))

plt.suptitle("Visualizing Relationships & Distributions among Features", fontsize=16)

plt.tight_layout();
```

```python
# heatmap of how correlated the features are with each other

fig, ax = plt.subplots(figsize=(4,4), dpi=200)

# correlation matrix
correlation = df.corr()

sns.heatmap(data=correlation, ax=ax,
           annot=True, cmap="PuBu")

fig.suptitle("Heatmap of Inter-Feature Correlation")
fig.tight_layout();
```

### Observation
1. **Fairly strong positive correlation** exist between
    - **Physical condition (`physical_score`)** & **Passing audio test**
    ---
1. **Fairly strong negative correlation** exist between
    - Participant **age (`physical_score`)** & **Passing audio test**
---
>_Heatmap supports previous analyses concluding that younger people with better physical condition are more likely to pass the audio test than older peopl._

### Visualizing the clusters for classification

```python
fig = plt.figure(dpi=200)
ax = fig.add_subplot(projection='3d')

x, y, z = df["age"], df["physical_score"], df["test_result"]

ax.scatter(x, y, z,
           marker="o", s=15, alpha=0.3,
          c=df["test_result"], cmap="coolwarm")

ax.view_init(elev=25, azim=190)

ax.set_xlabel("Age (years)")
ax.set_ylabel("Physical score")
ax.set_zlabel("Test result")

fig.tight_layout();
```

### Observation
>As the 3D scatterplot reveals, the dataset forms **2 distinct clusters** for the participants based on **result of audio test**.
>>Hence **classification algorithm** such as **_Logistic Regression_** should be able to learn and generalize well for newer examples later.
